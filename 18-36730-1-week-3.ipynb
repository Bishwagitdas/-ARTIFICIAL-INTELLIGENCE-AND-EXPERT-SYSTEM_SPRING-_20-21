Table driven agent

Table carwash agent(if robot see any dirty things try to wash this)

```python
table={('A','wash'):'Right',('A','Dirty'):'Need wash',('B','wash'):'Left',('B','Dirty'):'Need wash'}
percepts=[]    
def table_Carwash_agent(percept):
    print(percept)
    percepts.append(percept)
    action = lookup(percept,table)
    return action

def lookup(percept,table):
    for key in table:
        if key == percept:
            return table[key]

import random
Location = random.choice(['A','B'])
Condition = random.choice(['wash','Dirty'])

while True:
    action= table_Carwash_agent((Location,Condition))
    print(action)
    cmd = input()
    if(cmd == 'stop'): break
    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['wash','Dirty'])
    elif action== 'Left':
        Location = 'A'
        Condition = random.choice(['wash','Dirty'])
    else:
        Condition = 'wash'   
```

# Simple reflex agent

```python
def simple_Carwash_agent(percept):
    print(percept)
    if percept[1]=='wash':
        action = 'Need wash'
    elif percept[0]=='A':
        action = 'Right'
    elif percept[0]=='B':
        action = 'Left'
    return action

import random
Location = random.choice(['A','B'])
Condition = random.choice(['wash','Dirty'])

while True:
    action= simple_Carwash_agent((Location,Condition))
    print(action)
    cmd = input()
    if(cmd == 'stop'): break
    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['wash','Dirty'])
    elif action== 'Left':
        Location = 'A'
        Condition = random.choice(['wash','Dirty'])
    else:
        Condition = 'wash'

    
```

# Model based agent

```python
model = {'A':'Unknown', 'B':'Unknown'}
state = 'Bad'
action = 'Unknown'

def update_state(percept, model):
    global state, action
    if action=='Need wash':
        model[percept[0]]='wash'
    if model['A']=='wash' and model['B']=='wash':
        state = 'Good'
    else:
        state = 'Bad'
    return state
        

def model_based_Carwash_agent(percept):
    global state,action,model
    model[percept[0]]=percept[1]
    state = update_state(percept, model)
    print('Perception: '+str(percept)) 
    print('Model: '+str(model))   
    print('State: '+str(state))  
    if state == 'Good':
        action = 'Nothing to do'
    elif percept[1]=='Dirty':
        action = 'Need wash'
    elif percept[0]=='A':
        action = 'Right'
    elif percept[0]=='B':
        action = 'Left'
    return action


import random
Location = random.choice(['A','B'])
Condition= random.choice(['wash','Dirty'])

while True:
    action = model_based_Carwash_agent((Location,Condition))
    print('Action: '+ str(action))
    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['wash','Dirty'])
    elif action == 'Left':
        Location = 'A'
        Condition = random.choice(['wash','Dirty'])
    elif action == 'Need wash':
        Condition = 'wash'
    elif action == 'Nothing to do':
        break
        

    
```
