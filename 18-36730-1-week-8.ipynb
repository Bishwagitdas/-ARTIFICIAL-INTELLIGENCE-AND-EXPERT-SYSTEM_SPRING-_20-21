# Problem: Maximie f(x) = x^2 when x= 1 to 31

# Step 1: Initialize Population

```python
[i for i in range(4)]
```

```python
import random
def initPopulation(n,b):
    p={}
    for i in range(n):
        s=[]
        for j in range(b):
            s.append(random.choice([0,1]))    
        p[i]=s
    return p      

```

```python
population=initPopulation(6,5)
population
```

# Step 2: Parent Selection

Fitness Calculation

```python
[i for i in range(4,-1,-1)]
```

```python
def getFitness(p):
    f=[]
    for k in p.keys():
        d = 0
        b = len(p[k])
        for i in range(b-1,-1,-1):
            d=d+p[k][i]*2**(4-i)
        f.append(d**2)
    return f
```

```python
fitness = getFitness(population)
fitness
```

 probability

```python
total_fitness = sum(fitness)
prob = [round(f/total_fitness,3) for f in fitness]
prob
```

```python
def select_parent(n,fitness):
    total_fitness = sum(fitness)
    probabilities = [round(f/total_fitness,3) for f in fitness]
    parents = {}
    for i in range(n):
        r =random.choice(range(1,1001,1))/1000
        s = 0
        for j in range(len(probabilities)):
            s = s + probabilities[j]
            if(s>r):
                parents[i]=population[j]
                #print(population[j])
                break
    return parents
```

```python
parents = select_parent(4,fitness)
parents
```

```python
import random
def Crossover(parent1,parent2,point):
    p1,p2 = list(parent1),list(parent2) 
    for i in range(point,len(p1)):
        p1[i],p2[i] = p2[i],p1[i]       
    p1,p2 = ''.join(p1),''.join(p2)  
    return p1,p2
parent1 = '11001'    
parent2 = '11000'
print('Parent1:',parent1)
print('Parent2:',parent2)
point = random.randint(1,len(parent1))
print('Crossover Point:',point)
offspring1,offspring2 = Crossover(parent1,parent2,point)
print('Offspring1:',offspring1)        
print('Offspring2:',offspring2)
```
