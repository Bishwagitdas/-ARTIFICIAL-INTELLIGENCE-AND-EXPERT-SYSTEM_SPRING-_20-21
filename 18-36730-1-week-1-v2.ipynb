User-Defined Classes

```python
class Worker:
    def __init__(self, name, pay): 
        self.name = name 
        self.pay = pay
    def lastName(self):
        return self.name.split( )[-1] 
    def giveRaise(self, percent):
        self.pay *= (1.0 + percent)
```

```python
bishwagit = Worker('bishwagit das', 30000) 
joy = Worker('joy kumar', 20000) 
bishwagit.lastName( )
```

```python
joy.lastName( )
```

```python
bishwagit.giveRaise(.25)
bishwagit.pay
```
