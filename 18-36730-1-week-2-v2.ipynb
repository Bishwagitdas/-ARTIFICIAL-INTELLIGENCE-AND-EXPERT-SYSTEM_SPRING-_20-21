Dictionaries

```python
d1 = dict(Dhaka=10, Rajshahi=50)
d2 = {'Dhaka': 10, 'Rajshahi': 50}
d3 = dict(zip(['Dhaka', 'Rajshahi'], [10, 50]))
d4 = dict([('Dhaka', 10), ('Rajshahi', 50)])
d5 = dict({'Rajshahi': 50, 'Dhaka': 10})
print(d1)
print(d2)
print(d3)
print(d4)
print(d5) 
```

```python
d1 = {}
d1[0] = "Dhaka"
d1[1] = "Rajshahi"
print(d1)
```

```python
d1 = {'Dhaka': 10, 'Rajshahi': 50}
x = d1["Dhaka"]
y = d1.get("Rajshahi")
print(x)
print(y)
```
